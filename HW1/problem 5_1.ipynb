```python
import pandas as pd
```

```python
# import house_data dataset
original_dataset = pd.read_csv('kc_house_data.csv')
original_dataset.head()
```

```python
# ignore columns: id, date and zipcode
# also ignore price column because this is the response and what we are trying to predict
features = original_dataset.drop(columns=['id', 'date', 'zipcode', 'price'])
features.head()
```

1. For each feature, write code to compute the average value, the min and max values, as well as its variance.

```python
# for loop with each column of dataset that stores each stat as a variable
for feature in features:
    avg = features.mean()
    min = features.min()
    max = features.max()
    var = features.var()

# organize all variables into a table
stats = {
    'average value': avg,
    'min value': min,
    'max value': max,
    'variance': var
}

feature_data = pd.DataFrame(stats)
feature_data
```

Which features have the lowest and the highest average? Include the feature name and their average values for the features with the
lowest and highest average.

```python
max_avg = feature_data['average value'].max()
max_feature = feature_data['average value'].idxmax()

min_avg = feature_data['average value'].min()
min_feature = feature_data['average value'].idxmin()

print("Feature with the highest average was:", max_feature, "with an average of:", max_avg)
print("Feature with the lowest average was:", min_feature, "with an average of:", min_avg)
```

Which features have the lowest and the highest variance? Include the feature name and their variance values for the features with the
lowest and highest variance.

```python
max_var = feature_data['variance'].max()
max_feature1 = feature_data['variance'].idxmax()

min_var = feature_data['variance'].min()
min_feature1 = feature_data['variance'].idxmin()

print("Feature with the highest variance was:", max_feature1, "with an average of:", max_var)
print("Feature with the lowest variance was:", min_feature1, "with an average of:", min_var)
```
