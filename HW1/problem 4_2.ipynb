2. Generate at random 3 matrices of size 3 x 3 and fill each entry with a random integer chosen from -10 to 10. Use an existing
package to compute the inverse of each matrix if it exists. Include in your report:

- Each of the 3 matrices
- The inverses computed with the package
- Compute the product of each matrix with its inverse to check that you obtain the identity matrix.

```python
import numpy as np
```

```python
matrix1 = np.random.randint(-10 , 10, (3, 3), dtype="l")
matrix1
```

```python
matrix2 = np.random.randint(-10 , 10, (3, 3), dtype="l")
matrix2
```

```python
matrix3 = np.random.randint(-10 , 10, (3, 3), dtype="l")
matrix3
```

```python
# inverse of matrix1 using numpy's linalg package
inv1 = np.linalg.inv(matrix1)
inv1
```

```python
inv2 = np.linalg.inv(matrix2)
inv2
```

```python
inv3 = np.linalg.inv(matrix3)
inv3
```

```python
# compute the product of the inverse1 with matrix1 to check that identity matrix is obtained
inv1 @ matrix1
```

```python
inv2 @ matrix2
```

```python
inv3 @ matrix3
```
