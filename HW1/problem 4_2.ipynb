2. Generate at random 3 matrices of size 3 x 3 and fill each entry with a random integer chosen from -10 to 10. Use an existing
package to compute the inverse of each matrix if it exists. Include in your report:

- Each of the 3 matrices
- The inverses computed with the package
- Compute the product of each matrix with its inverse to check that you obtain the identity matrix.

```python
import numpy as np
```

```python
# generate 3 matrices of size 3x3 and fill with random integers from -10 to 10
matrices = []
for i in range(3):
    matrices.append(np.random.randint(-10, 10, (3, 3), dtype="l"))
matrices
```

```python
# find the inverse of each matrix using the numpy linalg package
inv_matrices = []
for i in matrices:
    inv_matrices.append(np.linalg.inv(i))
inv_matrices
```

```python
# compute the product of the inverse matrix and original matrix to double check that 
# the identity matrix is obtained
identity_matrices = []
for i in range(len(inv_matrices)):
    identity_matrices.append(matrices[i] @ inv_matrices[i])
identity_matrices
```
